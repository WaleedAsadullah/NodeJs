```python
# Import the library Selenium
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By
import time
from bs4 import BeautifulSoup
import re


# Make browser open in background
options = webdriver.ChromeOptions()
options.add_argument('headless')

# Create the webdriver object
browser = webdriver.Chrome(
    executable_path="C:\chromedriver_win32\chromedriver.exe",
options=options)
```

```python
url = ['https://www.google.com/maps/search/Pharmacies/@38.9104876,-77.0385897,14z/data=!3m1!4b1!4m7!2m6!3m5!2sCambria+Hotel+Washington,+D.C.+Convention+Center!3s0x89b7b7edb9b663c1:0x9a9e80ae4f8309da!4m2!1d-77.0234597!2d38.9089457?authuser=0&hl=en']
```

```python
for i in range(len(url)):
    browser.get(url[i])
    scrollable_elements = browser.find_elements(By.CLASS_NAME,"m6QErb")
    for scrollable_element in scrollable_elements:
        for i in range(1,2):
            time.sleep(1)
            try:
                browser.execute_script("arguments[0].scroll(0, arguments[0].scrollHeight);", scrollable_element)
            except:
                print("=")
```

```python
get_business_urls = browser.find_elements(By.TAG_NAME, 'a')
```

```python
page_source = browser.page_source
soup = BeautifulSoup(page_source, 'html.parser')
```

```python
soup.find_all('a')

business_links = []
for link in soup.find_all('a', attrs={'href': re.compile("^https://www.google.com/maps/place/")}):
    business_links.append(link.get('href'))
```

```python
business_links

for get_business_url in business_links:
    print(get_business_url)
    browser.get(get_business_url)
    
    business_page_source = browser.page_source
    business_soup = BeautifulSoup(business_page_source, 'html.parser')
    business_badges = business_soup.body.findAll('div', attrs={'class': 'Io6YTe'})
    for business_badge in business_badges:
        print(business_badge)
```
